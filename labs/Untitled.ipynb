```python
import pandas as pd
df = pd.read_csv("housing/boston_fixed.csv")
df.head()
```

```python
df.shape
```

```python
df.cov()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
print(scaler.fit(df.drop("CHAS",axis=1)))

df2 = scaler.transform(df.drop("CHAS",axis=1))
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df2)

print(pca.explained_variance_ratio_)

print(pca.singular_values_)
```

```python
print(pca.components_)
```

```python
df_pca = pca.transform(df2)
```

```python
scores = pd.DataFrame(df_pca,columns=["PC1","PC2"])
```

```python
%matplotlib inline
scores.plot.scatter(x="PC1",y="PC2",c=df["CHAS"].map({0: "Blue", 1: "Red"}),s=df["MEDV"])
```
